```python
##function check broken images path.
def img_verify(sub_testlist):
    _except = []
    for file in sub_testlist:
        try:
            img = Image.open(file)  # open the image file
            img.verify()  # verify that it is, in fact an image
        except (IOError, SyntaxError) as e:
            _except.append(file)
    return _except 
```
